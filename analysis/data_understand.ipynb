#### データ理解

```python
import sys
sys.path.append("../") # homeディレクトリをpathに追加

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
# import warnings
# warnings.simplefilter('error')
```

```python
df = pd.read_excel("../data/元データ/20241202_PSA_25+25Lmin_浦部加工.xlsx", index_col=1,
                   header=0, sheet_name="吸着データ（全塔）").drop(columns=["モード"])
plt.rcParams["font.size"] = 14

plt.figure(figsize=(25,15))
plt.subplot(311)
plt.plot(df["P-1絶対圧 [kPa]"], label="T1")
plt.plot(df["P-2絶対圧 [kPa]"], label="T2")
plt.plot(df["P-3絶対圧 [kPa]"], label="T3")
# plt.ylim([0,20])
# plt.xlim([0,0.8])
plt.title("絶対圧")
plt.grid()
plt.legend()

plt.subplot(312)
plt.plot(df["T-1-下C [℃]"], label="T1-下", c="tab:blue", linestyle="-")
plt.plot(df["T-1-中C [℃]"], label="T1-中", c="tab:blue", linestyle="--")
plt.plot(df["T-1-上C [℃]"], label="T1-上", c="tab:blue", linestyle=":")
plt.plot(df["T-2-下C [℃]"], label="T2-下", c="tab:orange", linestyle="-")
plt.plot(df["T-2-中C [℃]"], label="T2-中", c="tab:orange", linestyle="--")
plt.plot(df["T-2-上C [℃]"], label="T2-上", c="tab:orange", linestyle=":")
plt.plot(df["T-3-下C [℃]"], label="T3-下", c="tab:red", linestyle="-")
plt.plot(df["T-3-中C [℃]"], label="T3-中", c="tab:red", linestyle="--")
plt.plot(df["T-3-上C [℃]"], label="T3-上", c="tab:red", linestyle=":")
plt.title("温度（吸着層）")
plt.legend()
plt.grid()

plt.subplot(313)
plt.plot(df[["CO2流量  [L/min]", "N2流量 [L/min]"]], label=["co2", "n2"])
plt.legend()
plt.title("流量")
plt.grid()

plt.show()
```

```python
df = pd.read_excel("../data/20240624_ICTへの提供データ_PSA実験_編集_メイン.xlsx",
                   sheet_name="python実装用_吸着のみ_立ち上がり修正", index_col="time")
plt.rcParams["font.size"] = 14

plt.figure(figsize=(25,15))
plt.subplot(231)
plt.plot(df["Pressure"])
plt.title("絶対圧")
plt.grid()
plt.subplot(232)
plt.plot(df["temp_001_001"], label="temp_11", c="tab:blue", linestyle="-")
plt.plot(df["temp_002_001"], label="temp_21", c="tab:orange", linestyle="-")
plt.plot(df["temp_001_002"], label="temp_12", c="tab:blue", linestyle="--")
plt.plot(df["temp_002_002"], label="temp_22", c="tab:orange", linestyle="--")
plt.plot(df["temp_001_003"], label="temp_13", c="tab:blue", linestyle=":")
plt.plot(df["temp_002_003"], label="temp_23", c="tab:orange", linestyle=":")
plt.title("温度（吸着層）")
plt.legend()
plt.grid()
plt.subplot(233)
plt.plot(df["temp_inflow"])
plt.title("温度（入口ガス）")
plt.grid()
plt.subplot(234)
plt.plot(df["temp_out"])
plt.title("温度（外気温）")
plt.grid()
plt.subplot(235)
plt.plot(df[["flow_rate_co2", "flow_rate_n2"]], label=["co2", "n2"])
plt.legend()
plt.title("流量")
plt.grid()

plt.show()
```
