```python
output_filename = "../data/3塔データ.csv"

df = pd.read_excel("../data/元データ/20241202_PSA_25+25Lmin_浦部加工.xlsx",
                   sheet_name="吸着データ（全塔）", index_col=1).drop(columns=["モード"])
columns = {"CO2流量  [L/min]": "flow_rate_co2",
           "N2流量 [L/min]": "flow_rate_n2",
           "真空ポンプ排気量 [L/min]": "vaccume_rate",
           "T-1-上C [℃]": "T1_temp_3", "T-1-中C [℃]": "T1_temp_2", "T-1-下C [℃]": "T1_temp_1",
           "T-2-上C [℃]": "T2_temp_3", "T-2-中C [℃]": "T2_temp_2", "T-2-下C [℃]": "T2_temp_1",
           "T-3-上C [℃]": "T3_temp_3", "T-3-中C [℃]": "T3_temp_2", "T-3-下C [℃]": "T3_temp_1",
           "P-1絶対圧 [kPa]": "T1_press", "P-2絶対圧 [kPa]": "T2_press", "P-3絶対圧 [kPa]": "T3_press"
           }
df = df[columns.keys()].rename(columns=columns)
# 単位変換
df.loc[:, ["T1_press", "T2_press", "T3_press"]] = df.loc[:,["T1_press", "T2_press", "T3_press"]]/1000

# df.iloc[:2042,:].to_csv(output_filename)
df.to_csv(output_filename)
```
