```python
df = pd.read_excel("../data/20240624_ICTへの提供データ_PSA実験_編集_メイン.xlsx",
                   sheet_name="python実装用_すべて", index_col="time")
```

```python
plt.rcParams["font.size"] = 14

plt.figure(figsize=(25,15))
plt.subplot(231)
plt.plot(df["Pressure"])
plt.title("絶対圧")
plt.grid()
plt.subplot(232)
plt.plot(df["temp_11"], label="temp_11", c="tab:blue", linestyle="-")
plt.plot(df["temp_21"], label="temp_21", c="tab:orange", linestyle="-")
plt.plot(df["temp_12"], label="temp_12", c="tab:blue", linestyle="--")
plt.plot(df["temp_22"], label="temp_22", c="tab:orange", linestyle="--")
plt.plot(df["temp_13"], label="temp_13", c="tab:blue", linestyle=":")
plt.plot(df["temp_23"], label="temp_23", c="tab:orange", linestyle=":")
plt.title("温度（吸着層）")
plt.legend()
plt.grid()
plt.subplot(233)
plt.plot(df["temp_inflow"])
plt.title("温度（入口ガス）")
plt.grid()
plt.subplot(234)
plt.plot(df["temp_out"])
plt.title("温度（外気温）")
plt.grid()
plt.subplot(235)
plt.plot(df[["flow_rate_co2", "flow_rate_n2"]], label=["co2", "n2"])
plt.legend()
plt.title("流量")
plt.grid()

plt.show()
```
