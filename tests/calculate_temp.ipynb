```python
import CoolProp.CoolProp as CP  #pip install CoolProp
```

```python
# 温度と圧力を設定
T_C = 389.5  # 温度 [°C]
P = 4 * (10**6)  # 圧力 [Pa]

# 温度を絶対温度（ケルビン）に変換
T_K = T_C + 273.15

# 計算対象の流体を指定
fluids = 'water' # 他に、'co2'や'nitrogen'など

# 物性値を計算
density = CP.PropsSI('D', 'T', T_K, 'P', P, fluids)
viscosity = CP.PropsSI('V', 'T', T_K, 'P', P, fluids) * 1e6
thermal_conductivity = CP.PropsSI('L', 'T', T_K, 'P', P, fluids) 
specific_heat = CP.PropsSI('C', 'T', T_K, 'P', P, fluids) /1000

print(f"密度: {density} kg/m³")
print(f"粘度: {viscosity} μPa·s")
print(f"熱伝導度: {thermal_conductivity} W/(m·K)")
print(f"比熱容量: {specific_heat} J/(g·K)") #Cpのことです
```
