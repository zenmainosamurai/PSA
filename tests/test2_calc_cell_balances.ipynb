```python
cond_id = "2_20_test_num_section"
self = GasAdosorption_Breakthrough_simulator(cond_id)
```

- 状態変数の初期化

```python
variables = self.init_variables()
variables
```

- マテバラ

```python
material_output = {}
for stream in range(1, 1+self.num_str):
    material_output[stream] = {}
    material_output[stream][1] = self.calc_cell_material(stream=stream,
                                                         section=1,
                                                         variables=variables,
                                                         )
    material_output[stream][2] = self.calc_cell_material(stream=stream,
                                                         section=2,
                                                         variables=variables,
                                                         inflow_gas=material_output[stream][1])
    material_output[stream][3] = self.calc_cell_material(stream=stream,
                                                         section=3,
                                                         variables=variables,
                                                         inflow_gas=material_output[stream][2])
pprint.pprint(material_output)
```

- 熱バラ

```python
heat_output = {}
for stream in range(1, 1+self.num_str):
    heat_output[stream] = {}
    heat_output[stream][1] = self.calc_cell_heat(stream=stream,
                                                        section=1,
                                                        variables=variables,
                                                        material_output=material_output[stream][1])
    heat_output[stream][2] = self.calc_cell_heat(stream=stream,
                                                        section=2,
                                                        variables=variables,
                                                        material_output=material_output[stream][2],
                                                        heat_output=heat_output[stream][1],)
    heat_output[stream][3] = self.calc_cell_heat(stream=stream,
                                                        section=3,
                                                        variables=variables,
                                                        material_output=material_output[stream][3],
                                                        heat_output=heat_output[stream][2])
pprint.pprint(heat_output)
```

- 熱バラ（壁）

```python
heat_output[1+self.num_str] = {}
heat_output[1+self.num_str][1] = self.calc_cell_heat_wall(section=1,
                                               variables=variables,
                                               heat_output=heat_output[2][1])
heat_output[1+self.num_str][2] = self.calc_cell_heat_wall(section=2,
                                               variables=variables,
                                               heat_output=heat_output[2][2],
                                               heat_wall_output=heat_output[1+self.num_str][1])
heat_output[1+self.num_str][3] = self.calc_cell_heat_wall(section=3,
                                               variables=variables,
                                               heat_output=heat_output[2][3],
                                               heat_wall_output=heat_output[1+self.num_str][2])
pprint.pprint(heat_output[1+self.num_str])
```

- 熱バラ（上下蓋）

```python
heat_lid_output = {}
heat_lid_output["up"] = self.calc_cell_heat_lid(position="up",
                                                variables=variables,
                                                heat_output_dict=heat_output)
heat_lid_output["down"] = self.calc_cell_heat_lid(position="down",
                                                variables=variables,
                                                heat_output_dict=heat_output)
pprint.pprint(heat_lid_output)
```
