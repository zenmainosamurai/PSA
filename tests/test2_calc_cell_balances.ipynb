```python
variables = self.init_variables()
variables
```

- マテバラ

```python
material_output = {}
for stream in range(1, 1+self.num_str):
    material_output[stream] = {}
    material_output[stream][1] = self.calc_cell_material(stream=stream,
                                                         section=1,
                                                         variables=variables,
                                                         )
    material_output[stream][2] = self.calc_cell_material(stream=stream,
                                                         section=2,
                                                         variables=variables,
                                                         inflow_gas=material_output[stream][1])
    material_output[stream][3] = self.calc_cell_material(stream=stream,
                                                         section=3,
                                                         variables=variables,
                                                         inflow_gas=material_output[stream][2])
pprint.pprint(material_output)
```

- 熱バラ

```python
heat_output = {}
for stream in range(1, 1+self.num_str):
    heat_output[stream] = {}
    for section in range(1, 1+self.num_sec):
        heat_output[stream][section] = self.calc_cell_heat(stream=stream,
                                                           section=section,
                                                           variables=variables,
                                                           material_output=material_output[stream][section])
pprint.pprint(heat_output)
```

- 熱バラ（壁）

```python
heat_wall_output = {}
heat_wall_output[1] = self.calc_cell_heat_wall(section=1,
                                               variables=variables,
                                               heat_output=heat_output[2][1])
heat_wall_output[2] = self.calc_cell_heat_wall(section=2,
                                               variables=variables,
                                               heat_output=heat_output[2][2],
                                               heat_wall_output=heat_wall_output[1])
heat_wall_output[3] = self.calc_cell_heat_wall(section=3,
                                               variables=variables,
                                               heat_output=heat_output[2][3],
                                               heat_wall_output=heat_wall_output[2])
pprint.pprint(heat_wall_output)
```
